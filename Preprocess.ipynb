```python
df = pd.read_excel('/home/onyxia/work/Preprocess_WEF/wgidataset.xlsx')
```

```python
df = df[df['year'] == 2022]
```

```python
df_pivoted = df.pivot(index='code', columns='indicator', values=['estimate', 'stddev'])

```

```python
df_pivoted
```

```python
va = Voice and Accountability
pv Political Stability and Absence of Violence/Terrorism
ge = Government Effectiveness
rq = Regulatory Quality
rl = Rule of Law
cc = Control of Corruption
```

```python
rename_dict = {
    'va': 'Voice and Accountability',
    'pv': 'Political Stability and Absence of Violence/Terrorism',
    'ge': 'Government Effectiveness',
    'rq': 'Regulatory Quality',
    'rl': 'Rule of Law',
    'cc': 'Control of Corruption'
}

# Supposons que df_pivoted.columns = MultiIndex (value_type, indicator, year)

df_pivoted.rename(
    columns=lambda x: rename_dict[x] if isinstance(x, str) and x in rename_dict else x,
    level=1,  # On cible le niveau des indicateurs
    inplace=True
)
```

```python
df_pivoted
```

```python
df_pivoted.to_csv('WGI_SCORES.csv', index=True)
```
